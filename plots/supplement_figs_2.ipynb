```python
def simple_linear_regression(x, y):
    """
    Perform simple linear regression and return slope, intercept and R².
    """
    x = np.array(x)
    y = np.array(y)
    
    # Calculate means
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    # Calculate slope
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean) ** 2)
    slope = numerator / denominator
    
    # Calculate intercept
    intercept = y_mean - (slope * x_mean)
    
    # Calculate R²
    y_pred = slope * x + intercept
    r_squared = 1 - (np.sum((y - y_pred) ** 2) / np.sum((y - y_mean) ** 2))
    
    return slope, intercept, r_squared

slope_180ns, interception_180ns, r_squared_180ns = simple_linear_regression(nw_diag_inv[3:7], l2_norm_mean_180ns[3:7])
slope_360ns, interception_360ns, r_squared_360ns = simple_linear_regression(nw_diag_inv[3:7], l2_norm_mean_360ns[3:7])
slope_540ns, interception_540ns, r_squared_540ns = simple_linear_regression(nw_diag_inv[3:7], l2_norm_mean_540ns[3:7])
```
